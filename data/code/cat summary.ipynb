```python
import wikipedia
import requests
from bs4 import BeautifulSoup
import time
import re

class Cats:
    def __init__(self, name, summary):
    #def __init__(self, name, img_str, wiki_link, summary):
        self.name = name
        #self.img_str = img_str
        #self.wiki_link = wiki_link
        self.summary = summary

html = requests.get('https://en.wikipedia.org/wiki/List_of_cat_breeds')

#turn the HTML into a beautiful soup text object
b = BeautifulSoup(html.text, 'lxml')

cats = list()

# get to the table on the page
for i in b.find_all(name='table', class_='wikitable sortable'):
    for j in i.find_all(name='tr'):
        #for imgs, but we can probably append with kevin's for imgs
        #for l in j.find_all(name='td'):
        #    for m in l.find_all(name='div'):
        #        for n in m.find_all(name='div'):
        #            for o in n.find_all('a'):
        #                for img in o.find_all('img'):
        #                    cat_img.append(img['src'])
        
        
        #get the summary of each breed
        for k in j.find_all(name='th'):
            # get within that cell to just get the words
            for link in k.find_all('a', href=True):
                if(re.match('^#', link['href'])):
                    continue
                else:
                    #cat_name = link['title']
                    #cat_summary = wikipedia.WikipediaPage(link['title']).summary
                    cat = Cats(link['title'], wikipedia.WikipediaPage(link['title']).summary)   
                    #cat = Cats(cat_name, cat_summary)
                    cats.append(cat)
                    time.sleep(1)
```

```python
for cat in cats:
    print(cat.name)
    print(cat.summary)
    print('\n')
```
